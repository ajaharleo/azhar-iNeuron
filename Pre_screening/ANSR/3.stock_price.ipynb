3.You have a dataset named stock_prices.csv that contains daily stock prices of a company. The dataset has columns date and closing_price. Write a Python script to read the dataset, convert the date column to a datetime object, and then perform a time series analysis by plotting the closing price over time with a rolling 30-day moving average.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('stock_prices.csv')
```

```python
df.head()
```

```python
df.dtypes
```

```python
len(df)
```

```python
# date column data type is object so converting it to date type
df['date'] = pd.to_datetime(df['date'])
```

```python
# we need a rolling average over time so need to sort the table by date
df.sort_values(by='date',inplace = True)
```

```python
# 30 day rolling average
rolling_avg = df['closing_price'].rolling(window=10,min_periods=1).mean().round(2)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['date'], df['closing_price'], label='Closing Price', color='blue')
plt.plot(df['date'], rolling_avg, label='30-Day Avg', color='red')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Stock Prices Over Time with 30-Day Moving Average (Rounded to 2 Decimal Points)')
plt.legend()
plt.grid(True)
```
